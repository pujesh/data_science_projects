### Objective and Data Description

#### Overview
Banks provide term deposit to their customers and to attract customers to the feature, there are various marketing campaigns conducted. One way is a direct marketing through phone calls. Based on the history of the customer and the type of customers, we can predict on the likeliness of a customer to subscribe the term deposit. This prediction helps us to focus on those clients first and give high priority during the marketing campaign. 

#### Objective
**Predict the likeliness of a customer to subscribe a term deposit.**

#### About the Dataset
This dataset is being used from the Kaggle website and the URL to the dataset is: [Bank Term Deposit](https://www.kaggle.com/krantiswalke/bankfullcsv)

It has 16 fields and the description of each field of the dataset are: 
1. **AGE** - Age of the customer.
2. **JOB** - Type of job of the customer.
3. **MARITAL** - Marital status of the customer. 
4. **EDUCATION** - Education level of the customer.
5. **DEFAULT** - If the customer has credit in default. 
6. **BALANCE** - Average yearly balance of the customer.
7. **HOUSING** - If the customer has housing loan. 
8. **LOAN** - If the customer has personal loan. 
9. **DAY** - Last contact day of the month.
10. **CONTACT** - Contact communication type. 
11. **MONTH** - Last contact month of the year.
12. **DURATION** - Last contact duration (in seconds). 
13. **CAMPAIGN** - Number of contacts performed during this campaign for the customer. 
14. **PDAYS** - Number of days that passed by after the client was last contacted from a previous campaign. 
15. **PREVIOUS** - Number of contacts performed in the previous campaign. 
16. **POUTCOME** - Outcome of the previous marketing campaign. 
17. **TARGET** - Whether the client subscribed a term deposit.

```python
# All the required libraries are imported here.
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.ensemble import ExtraTreesClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
#Describing the size, shape and dimension of the dataset.
bank_data = pd.read_csv('bank_full.csv') 
print("The total size of the dataset is {} bytes.".format(bank_data.size))
print("There are {} number of records and {} number of fields with a {} shape.".format(bank_data.shape[0], bank_data.shape[1], bank_data.shape))
print("The dataset has a {} dimensional structure.".format(bank_data.ndim))
```

```python
print("The top 5 records of the dataset are: \n{}".format(bank_data.head()))
```

```python
#Listing the unique values for all the variables of object type.
for data_keys in bank_data.keys():
    if bank_data[data_keys].dtype == 'object':
        print("The unique value of {} variables are : {}".format(data_keys, list(bank_data[data_keys].unique())))
```

```python
#Each categorical variables of object types are replaced with numberical numberes
for data_keys in bank_data.keys():
    if bank_data[data_keys].dtype == 'object':
        print("{}:".format(data_keys))
        print("Before values: {}".format(bank_data[data_keys].unique()))
        uniq_value = list(bank_data[data_keys].unique())
        bank_data[data_keys] = bank_data[data_keys].replace(uniq_value, list(np.arange(len(uniq_value))))
        print("After values: {}\n".format(list(bank_data[data_keys].unique())))
```

### Algorithm Implementation and Statistical Tests 

```python
#Considering all the features for classification. 
pred_var = bank_data.iloc[:, 0:15]
targ_var = bank_data['Target']

X_train, X_test, y_train, y_test = train_test_split(pred_var, targ_var, test_size=0.25)

# Using the Standard Scaler for standardization of our dataset.
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

nb = GaussianNB()
nb.fit(X_train, y_train)

y_pred = nb.predict(X_test)

print("The accuracy score of the Gaussian Naive Bayes is {}".format(nb.score(X_test, y_test) * 100))
```

```python
#Creating confusion matrix and classification report on the prediction.
y_pred = nb.predict(X_test)
matrix_data = pd.DataFrame(confusion_matrix(y_test, y_pred), index=['F', 'T'], columns=['F', 'T'])
print("Confusion Matrix:\n",matrix_data)

print(classification_report(y_test,np.ravel(y_pred)))
```

We are able to gain a pretty high accuracy using the Gaussian Naive Bayes algorithm. Since the target variable predicts a boolean type, where it determines if a customer has subscribed the term deposit that was provided by the bank, Bernoulli Naive Bayes might not be able to predict with good accuracy score. This is because, Bernoulli Naive Bayes works well when the predictor features are in boolean type.

Let us see how Bernoulli Naive Bayes can predict for the given dataset.

```python
#Considering all the features for classification. 
pred_var = bank_data.iloc[:, 0:15]
targ_var = bank_data['Target']

X_train, X_test, y_train, y_test = train_test_split(pred_var, targ_var, test_size=0.25)

# Using the Standard Scaler for standardization of our dataset.
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

nb = BernoulliNB()
nb.fit(X_train, y_train)

y_pred = nb.predict(X_test)

print("The accuracy score of the Bernoulli Naive Bayes is {}".format(nb.score(X_test, y_test) * 100))
```

```python
#Creating confusion matrix and classification report on the prediction.
y_pred = nb.predict(X_test)
matrix_data = pd.DataFrame(confusion_matrix(y_test, y_pred), index=['F', 'T'], columns=['F', 'T'])
print("Confusion Matrix:\n",matrix_data)

print(classification_report(y_test,np.ravel(y_pred)))
```

Here, the Bernoulli was able to predict better than the Gaussian Naive Bayes classifier slightly. The accuracy is similar and any of the algorithm can be used.

### Performance Improvement

```python
#Finding the features that has negative value. Features having negative values cannot be used for finding important featuers. We will be removing these features from our algorithm
print("List of predictor features that contains negative values:")
for data_keys in bank_data.keys():
    if any(bank_data[data_keys] < 0):
        print(data_keys)
```

```python
#Considering all of the features to find the top features and removing pdays and balance since it contains negative values.
X = bank_data[['age', 'job', 'marital', 'education', 'default', 'housing', 'loan',\
                      'contact', 'day', 'month', 'duration', 'campaign', 'previous', 'poutcome']]
y = bank_data.iloc[:, -1]

select_k_best = SelectKBest(score_func=chi2, k=10)

best_result = select_k_best.fit(X, y)

scores = pd.DataFrame(best_result.scores_)
columns = pd.DataFrame(X.columns)

feature_scores = pd.concat([columns, scores], axis=1)
feature_scores.columns = ['Field', 'Score']

print("Listing the top 5 features : \n{}".format(feature_scores.nlargest(10, 'Score')))
```
